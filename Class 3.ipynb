```python
# Simplified implementation (relying on built-in data sturctures)

class Queue:
    def __init__(self):
        self.items = []

    def enqueue(self, value):
        return self.items.insert(0, value)

    def dequeue(self):
        return self.items.pop()

    #nice to have methods
    
    def size(self):
        return len(self.items)

    def peek(self):
        return self.items[len(self.items)-1]

    def is_empty(self):
        return self.items == [] 
```

```python
class SinglyLinkedList:
    class __Node:
        def __init__(self, data):
            self.data = data
            self.next = None

    def __init__(self):
        self.head = None
        # what would happen if we also maintained self.tail?

    def append(self, value):
        # append adds a new element to our list as teh tail (the last element in the list)
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next: # while current.next != None
                current = current.next
            current.next = new_node

    def insert(self, index, value):
        new_node = self.__Node(value)
        if not self.head:
            self.head = new_node
        else:
            if index == 0:
                new_node.next = self.head
                self.head = new_node
            else:
                current = self.head
                count = 0
                prev = None
                while current.next and count != index:
                    prev = current
                    current = current.next
                    count += 1
                if count == index:
                    prev.next = new_node
                    new_node.next = current
                else:
                    current.next = new_node
    
    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % repr(current.data)
            current = current.next
            while current:
                out += ", %s" % repr(current.data)
                current = current.next
        out += "]"
        return out

```
